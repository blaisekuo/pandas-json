```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import datetime as dt
```

```python
data = pd.read_json("fish1979-2018.json", typ='series')

print(data)
```

```python
t = len(data['DATA'])
rangt = range(t)
```

```python
firstyear = 1979
lastyear = 2018

plt.figure(figsize=(20, 12))
plt.xlabel('day of year')
plt.ylabel("fish count")
plt.title("Fish counts from " + str(firstyear) + "-" + str(lastyear) + " seasons",fontsize=15)

fishcount = {}
dayofyear = {}



for year in range(firstyear,lastyear+1):
    
    fishcountarray = []
    dayofyeararray = []
    for i in rangt:
        if data['DATA'][i][0] == str(year):
            fishcountarray.append(int(data['DATA'][i][2]))
            dayofyeararray.append(dt.datetime.strptime(data['DATA'][i][1], '%B, %d %Y %H:%M:%S').timetuple().tm_yday)
    
    fishcount.update({str(year): fishcountarray})
    dayofyear.update({str(year): dayofyeararray})
    
for year in range(firstyear,lastyear+1):
    #print(fishcount[str(year)])
    #plt.plot(dayofyear[k],fishcount[k],label= str(year))
    plt.plot(dayofyear[str(year)],fishcount[str(year)],label= str(year))
    
#plt.ylim(0, 80000)
plt.yticks(np.arange(0,250000,10000))
plt.legend(loc='best')
```
